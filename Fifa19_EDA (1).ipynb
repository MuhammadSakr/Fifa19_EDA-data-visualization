```python
df
```

# Data Analysis

# 

```python
def labels(x,y,title) :
    plt.xlabel(x,size=13)
    plt.ylabel(y,size=13)
    plt.title(title,size=15)
```

```python
plt.hist(df['value']);
labels('value','count','Players Values')
```

```python
df.value.describe()
```

# 

## 1) Who are the heighst valued players?

```python
values=df.sort_values(by=['value'],ascending=False).head(50).sort_values(by='value')
plt.figure(figsize=(20,10))
labels('Value ($)','Name','Heighst 50 Valued Players')
plt.barh(values['name'],values['value']);
```

# 



## 2) What are the best clubs?

```python
# get main players percentage to the whole team players
11/df.club.value_counts().mean()
```

```python
# clubs dataframe to see analysis about clubs related to players overall powers
clubs=df.groupby(by='club', as_index=False)['potential'].quantile(0.6)
# We used quantile of 0.6 cuz the main players are less than the bench (with main-high powers)
clubs=clubs.sort_values('potential',ascending=False)
clubs=clubs.head(50).sort_values(by='potential')
```

```python
plt.figure(figsize=(20,10))
plt.barh(clubs['club'],clubs['potential'])
labels('Potential_power avg (%)','Club','Best 50 club');
```

# 

## 3) What are most common players nationalities?

```python
nationality=pd.DataFrame(df.nationality.value_counts()).head(50).sort_values(by='nationality')
nationality=nationality.rename(columns={'nationality':'count'})
nationality['nationality']=nationality.index
```

```python
plt.figure(figsize=(20,10))
plt.barh(nationality['nationality'],nationality['count'])
# plt.xticks(rotation=90)
labels('Count','Nationality','Most common players nationalities')
```

# 

## 4) Which clubs pay more wages?

```python
# clubs dataframe to see analysis about clubs related to players overall powers
clubs=df.groupby(by='club', as_index=False)['wage'].mean()
# We used quantile of 0.6 cuz the main players are less than the bench
clubs=clubs.sort_values('wage',ascending=False)
clubs=clubs.head(25).sort_values(by='wage')
```

```python
plt.figure(figsize=(20,10))
plt.barh(clubs['club'],clubs['wage'])
labels('Wages avg ($)','Club','Which clubs pay more wages?');
```

## 12) What is the correlation between player value & potential_power?

```python
plt.figure(figsize=(20,10))
sb.regplot(data = df, x='potential',y='value',truncate=False, x_jitter=0.3,scatter_kws={'alpha':0.3});
labels('Potential power (%)','Value ($)','Player value - potential_power relation')
```

* We find a positive relationship between player value & potential power.

* Notice the exponential growth in player value by increasing potential power.

# 

## 13) Does potential player power matter in his international_reputation?

```python
plt.figure(figsize=(20,10))
sb.regplot(data = df, x='potential',y='international_reputation',truncate=False, x_jitter=0.3,scatter_kws={'alpha':0.3});
labels('Potential_power (%)','International_reputation (out of 5)','Player international_reputation - potential power relation');
```

* We find a positive correlation between international_reputation and potential player power.

# 

## 14) What is the relation between Player weight - skill_moves?

```python
plt.figure(figsize=(20,10))
sb.regplot(data = df, x='weight',y='skill_moves',truncate=False, x_jitter=0.3,scatter_kws={'alpha':0.3});
labels('Weight (pound)','Skill_moves (out of 5)','Player weight - skill_moves relation');
```

* Negative correlation
* Notice most heavy weight players aren't skillfull.
* Most skilled players' weights are almost between 140-160.

# 

## 15) Which positions require more skill_moves rate?

```python
plt.figure(figsize=(20,10))
order = df.groupby(["position"])["skill_moves"].mean().sort_values(ascending=False).index
sb.pointplot(data=df,x='position',y='skill_moves',color=sb.color_palette()[0],order=order)
labels('Position','Skill_moves avg (out of 5)','Which positions require more skill_moves rate?');
plt.savefig('Which positions require more skill_moves rate 3');
```

## 16) What is the relationship between player age, potential power & international_reputation?

```python
plt.figure(figsize=(20,10))
plt.scatter(data = df, x = 'potential', y = 'age', c = 'international_reputation',cmap='Paired')
plt.colorbar(label='international_reputation (out of 5)')

m, b = np.polyfit(df['potential'],df['age'], 1)
#add linear regression line to scatterplot 
plt.plot(df['potential'], m*df['potential']+b,color='r')

labels('Potential_power (%)','Age','What is the relationship between player_age, potential_power & international_reputation?');
```

* Negative correlation between age & potential_power.

* Players with high international_reputation are concentrated in ages(25-35) & powers above 80.

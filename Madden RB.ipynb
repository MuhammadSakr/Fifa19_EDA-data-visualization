```python
# 13/2 changes *********************************
# Deleted F1 , Reformulate base projections


# players['F1/Total number of team games played']=np.nan
players['Base Floor Projection']=np.nan
players['Base Ceiling Projection']=np.nan
for j in range(len(players)) :
#     players['F1/Total number of team games played'].values[j]=(players['player_games'].values[j]/players['team_games'].max())*players['player_avg_fpts'].values[j]
#     players['Base Floor Projection'].values[j]=players['F1/Total number of team games played'].values[j]-(players.avg_std.values[j]*0.68)
#     players['Base Ceiling Projection'].values[j]=players['F1/Total number of team games played'].values[j]+(players.player_avg_fpts.values[j]*0.68)

    
    players['Base Floor Projection'].values[j]=(players['player_avg_fpts'].values[j]*players['Percentage of games played in'].values[j])-(players.avg_std.values[j]*0.68)
    players['Base Ceiling Projection'].values[j]=(players['player_avg_fpts'].values[j]*players['Percentage of games played in'].values[j])+(players.avg_std.values[j]*0.68)
```

```python
players
```

```python
# 13/2 changes **********************************
# Added adjusted




for i in range(len(players)) :
    floorn='Floor VS '+str(players.Team.values[i])
    ceilingn='Ceiling VS '+str(players.Team.values[i])
    adjusted="Adjusted projection VS "+str(players.Team.values[i])
    
    players[floorn]=np.nan
    players[ceilingn]=np.nan
    players[adjusted]=np.nan
    for j in range(len(players)) :
        players[floorn].values[j]=players['adjusted_fpts_vs_team'].values[i]-(players.avg_std.values[j]*0.68)
        players[ceilingn].values[j]=players['adjusted_fpts_vs_team'].values[i]+(players.player_avg_fpts.values[j]*0.68)
        players[adjusted].values[j]=players['player_avg_fpts'].values[j]*(players['Position FPTS per game'].values[i]/players['AVG FPTS ALLOWED for all teams vs Position'].values[j])
players
```
